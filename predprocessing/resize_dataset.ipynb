```python
import pandas as pd
import numpy as np
from wand.image import Image
import os
from tqdm import tqdm
```

```python
df=pd.read_excel('razm.xlsx')
```

```python
file_list = []
for img in tqdm(os.listdir("/Итоговый датасет")):
    file_list.append(img)
    ny = Image(filename=f'Итоговый датасет/{img}')
    ny.resize(256, 256)
    ny_convert = ny.convert('jpg')
    ny_convert.save(filename=f'img/{os.path.splitext(img)[0]}.jpg')
```

```python
def change_ext(file):
    return f'img/{os.path.splitext(img)[0]}.jpg'


df = df[df['Названия'].isin(file_list)]
df['Названия'] = df['Названия'].map(change_ext)
```
