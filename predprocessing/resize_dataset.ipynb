```python
import pandas as pd
import numpy as np
from wand.image import Image
import os
from tqdm import tqdm
import json
```

```python
df=pd.read_excel('razm.xlsx')
```

```python
# with open('files.json') as f:
#     file_list = json.loads(f.read())

for img in tqdm(df['Названия']):
    ny = Image(filename=f'Итоговый датасет/{img}')
    size=min(ny.height,ny.width)
    ny.crop(width=size, height=size, gravity='center')
    ny.resize(256, 256)
    ny_convert = ny.convert('jpg')
    ny_convert.save(filename=f'img/{img}.jpg')
```

```python
def change_ext(file):
    return f'img/{file}.jpg'


# df = df[df['Названия'].isin(file_list)]
df['Названия'] = df['Названия'].map(change_ext)
```

```python
df['День'] = df['Время суток'].map(lambda a: 1 if a == 1 else 0)
df['Ночь'] = df['Время суток'].map(lambda a: 1 if a == 2 else 0)
df['Рассвет/закат'] = df['Время суток'].map(lambda a: 1 if a == 3 else 0)

df['Зима'] = df['Время года'].map(lambda a: 1 if a == 1 else 0)
df['Весна'] = df['Время года'].map(lambda a: 1 if a == 2 else 0)
df['Лето'] = df['Время года'].map(lambda a: 1 if a == 3 else 0)
df['Осень'] = df['Время года'].map(lambda a: 1 if a == 4 else 0)

df['Лес'] = df['Местность'].map(lambda a: 1 if a == 1 else 0)
df['Город'] = df['Местность'].map(lambda a: 1 if a == 2 else 0)

df = df.drop(columns=['Время суток', 'Время года', 'Местность'])
```

```python
rng = np.random.RandomState()

train = df.sample(frac=0.8, random_state=rng)
test = df.loc[~df.index.isin(train.index)]
```

```python
train.to_csv('train.csv',index=None)
test.to_csv('test.csv',index=None)
```
